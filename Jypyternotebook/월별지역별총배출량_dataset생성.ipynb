## base_date 변환해서 year, month 컬럼 생성

```python
df['base_date'] = pd.to_datetime(df['base_date'])
```

```python
df.info()
```

```python
df['year_date'] = df['base_date'].dt.year
df['month_date'] = df['base_date'].dt.month
```

```python
df
```

## 목적에 맞는 필요 변수 선택하고 진행

```python
jejudf = df.loc[:, ['year_date', 'month_date', 'emd_nm', 'em_g']]
gd = jejudf.groupby(['year_date', 'month_date', 'emd_nm']).sum()
gd
```

```python
gd.to_csv("./월별지역별배출량.csv", encoding='CP949')
```

```python
jejudf2 = df.loc[:, ['year_date', 'month_date', 'city', 'emd_nm', 'em_g']]
gdc = jejudf2.groupby(['year_date', 'month_date', 'city']).sum()
gdc
```

```python
gdc.to_csv("./월별시별배출량.csv", encoding='CP949')
```

## g 단위 -> kg단위 수정
일자별 기록된 음식물쓰레기 배출량(g)은 sum()으로 합산시 단위가 커지는 문제가 발생함. 따라서 kg단위로 수정함.

```python
gdc1 = gdc.copy()
gdc1['em_g'] = round(gdc1['em_g'] / 1000, 0)
gdc1 = gdc1.rename({'em_g':'em_kg'}, axis=1)
gdc1
```

```python
gd1 = gd.copy()
gd1['em_g'] = round(gd1['em_g'] / 1000, 0)
gd1 = gd1.rename({'em_g':'em_kg'}, axis=1)
gd1
```

## int64로 배출량 단위 변환
다른 데이터 셋과 통합하는 경우를 생각해서 변수 타입을 int64로 통일

```python
# 변수 타입 int64로 통합
gd1['em_kg'] = gd1['em_kg'].astype('int64')
gdc1['em_kg'] = gdc1['em_kg'].astype('int64')
```
