## 2. 년도/월/일 열 분리

```python
df['year_date'] = df['base_date'].str.split('-').str[0]
df['month_date'] = df['base_date'].str.split('-').str[1]
df['day_date'] = df['base_date'].str.split('-').str[2]
df
```

```python
df[['year_date', 'month_date', 'day_date']] = df[['year_date', 'month_date', 'day_date']].apply(pd.to_numeric)
df
```

```python
data = df.copy()
```

## 3. groupby

```python
grouped = data['pay_amt'].groupby([data['year_date'],data['month_date'],data['emd_nm']])
grouped.sum()
```

```python
region_pay = pd.DataFrame(grouped.sum()).reset_index()
region_pay
```

```python
region_pay[(region_pay['year_date']==2021) & (region_pay['month_date']==6)]
```
