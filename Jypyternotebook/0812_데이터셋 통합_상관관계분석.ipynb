```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1) data1 : 음식물쓰레기 데이터

```python
data1 = pd.read_csv('/content/drive/MyDrive/21 세미프로젝트_2차 (1)/데이터/01_음식물쓰레기_기본수정.csv',encoding='cp949')
data1
```

```python
data1['year_date'] = data1['base_date'].str.split('-').str[0]
data1['month_date'] = data1['base_date'].str.split('-').str[1]
data1['day_date'] = data1['base_date'].str.split('-').str[2]
data1
```

```python
data1[['year_date', 'month_date', 'day_date']] = data1[['year_date', 'month_date', 'day_date']].apply(pd.to_numeric)
data1
```

```python
data1.info()
```

### 3) data3 : 거주인구 데이터

```python
data3 = pd.read_csv('/content/drive/MyDrive/21 세미프로젝트_2차/원본데이터/01_제공데이터_update(210811)/03_거주인구_RESIDENT_POP.CSV',encoding='cp949')
data3
```

```python
grouped = data3[['resid_reg_pop','foreign_pop','total_pop']].groupby([data3['base_year'],data3['base_month'],data3['emd_cd'],data3['emd_nm']])
data3_2 = pd.DataFrame(grouped.sum()).reset_index()
data3_2
```

### 4) data4 : 카드소비 데이터

```python
data4 = pd.read_csv('/content/drive/MyDrive/21 세미프로젝트_2차/원본데이터/01_제공데이터_update(210811)/04_음식관련 카드소비_CARD_SPENDING.CSV',encoding='cp949')
data4
```

```python
# '알수없음'버리기
# data4_2 = data4.copy()
data4_2 = data4[data4['emd_nm']!='알수없음']
data4_2
```

```python
data4_2.info()
```

```python
data4_2[['emd_cd']] = data4_2[['emd_cd']].apply(pd.to_numeric)
data4_2.info()
```

#### 1) 마트/슈퍼마켓 : market

```python
market = data4_2[data4_2['mct_cat_cd']=='A00200']
market
```

#### 2) 한식 : korean

```python
korean = data4_2[data4_2['mct_cat_cd']=='C00100']
```

#### 3) 양식~부페 : rest2

```python
rest = data4_2[(data4_2['mct_cat_cd'] == 'C00200') | (data4_2['mct_cat_cd'] == 'C00300') | (data4_2['mct_cat_cd'] == 'C00400') | (data4_2['mct_cat_cd'] == 'C00500') | (data4_2['mct_cat_cd'] == 'C00700')]
rest
```

```python
grouped = rest[['use_cnt','use_amt']].groupby([rest['base_date'],rest['emd_cd']])
rest2 = pd.DataFrame(grouped.sum()).reset_index()
rest2
```

#### 4) 주점 및 주류판매~농축수산물

```python
ect = data4_2[(data4_2['mct_cat_cd'] == 'C00900') | (data4_2['mct_cat_cd'] == 'C01000') | (data4_2['mct_cat_cd'] == 'C01200')]
grouped = ect[['use_cnt','use_amt']].groupby([ect['base_date'],ect['emd_cd']])
ect2 = pd.DataFrame(grouped.sum()).reset_index()
ect2
```

#### 5) 배달 : deli

```python
deli = data4_2[data4_2['mct_cat_cd']=='C01400']
deli
```

## 데이터셋 통합하기

### 1) data1 + total

```python
data3_2.columns
```

```python
data3_2.columns = ['year_date', 'month_date', 'emd_cd', 'emd_nm', 'resid_reg_pop', 'foreign_pop', 'total_pop']
data3_2
```

```python
final = pd.merge(data1, data3_2, left_on=['year_date','month_date','emd_cd'], right_on=['year_date','month_date','emd_cd'], how='outer')
final
```

```python
final[final['emd_nm_y']=='삼도1동']
```

```python
final['emd_nm'] = np.where(pd.isna(final['emd_nm_x']) == False, final['emd_nm_x'], final['emd_nm_y'])
final
```

```python
final = final.drop(['emd_nm_x','emd_nm_y'], axis=1)
final
```

```python
final['emd_nm'].unique()
```

### 2) final + market

```python
market.columns
```

```python
market.columns = ['base_date', 'city', 'emd_cd', 'emd_nm', 'mct_cat_cd', 'mct_cat_nm',
       'm_use_cnt', 'm_use_amt']
```

```python
final2 = pd.merge(final, market, left_on=['base_date', 'city', 'emd_cd'], right_on=['base_date', 'city', 'emd_cd'], how='outer')
final2
```

```python
final2['emd_nm'] = np.where(pd.isna(final2['emd_nm_x']) == False, final2['emd_nm_x'], final2['emd_nm_y'])
final2
```

```python
final2 = final2.drop(['emd_nm_x','emd_nm_y','mct_cat_cd','mct_cat_nm'], axis=1)
final2
```

### 3) final2 + korean

```python
korean.columns
```

```python
korean.columns = ['base_date', 'city', 'emd_cd', 'emd_nm', 'mct_cat_cd', 'mct_cat_nm',
       'k_use_cnt', 'k_use_amt']
```

```python
final3 = pd.merge(final2, korean, left_on=['base_date', 'city', 'emd_cd'], right_on=['base_date', 'city', 'emd_cd'], how='outer')
```

```python
final3['emd_nm'] = np.where(pd.isna(final3['emd_nm_x']) == False, final3['emd_nm_x'], final3['emd_nm_y'])
```

```python
final3 = final3.drop(['emd_nm_x','emd_nm_y','mct_cat_cd','mct_cat_nm'], axis=1)
final3
```

```python
rest2.columns
```
