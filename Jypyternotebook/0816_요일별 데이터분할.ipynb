## 1. 데이터셋 불러오기

```python
df = pd.read_csv('/content/drive/MyDrive/21 세미프로젝트_2차 (1)/데이터/01_음식물쓰레기_기본수정.csv',encoding='cp949')
df
```

## 2. datetime(날짜타입) 변수 설정

```python
df['date'] = pd.to_datetime(df['base_date'])
df
```

```python
df['year_date'] = df['base_date'].str.split('-').str[0]
df['month_date'] = df['base_date'].str.split('-').str[1]

df[['year_date', 'month_date']] = df[['year_date', 'month_date']].apply(pd.to_numeric)
df
```

```python
df0 = df[(df['year_date']==2021) & (df['month_date']==6)]
df0
```

```python
df0['week'] = df0['date'].dt.day_name()
df0
```

## 3. groupby 사용

```python
grouped1 = df0[['em_g','pay_amt']].groupby([df0['emd_nm'],df0['week']])
df1 = pd.DataFrame(grouped1.sum()).reset_index()
df1
```

```python
df1['em_kg'] = df1['em_g'] / 1000
df1
```
